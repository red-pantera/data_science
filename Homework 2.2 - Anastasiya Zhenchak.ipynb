Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
# 1ий спосіб
sex = input('Введіть стать спортсмена: ')
age = float(input('Введіть вік спортсмена: '))

valid_age = age in range(0, 101)
valid_sex = sex in ('f', 'm')

if valid_age == False:
    print('Вік введено невірно. Спробуйте ще раз')
if valid_sex == False:
    print('Стать введено невірно. Спробуйте ще раз')

if valid_age & valid_sex:
    if ((sex == 'f') & (age > 16)) or ((sex == 'm') & (age > 18)):
        print('Вітаю! Ви можете брати участь у змаганні')
    else:
        print('Вибачте, на жаль, ви не можете брати участь у змаганні')
```

```python
# 2ий спосіб з try... except
class SexError(Exception):
    pass

try:
    sex = input('Введіть стать спортсмена (m/f): ')
    age = float(input('Введіть вік спортсмена: '))
    
    if sex not in ('m','f'):
        raise SexError     
    if (sex == 'm') &  (16 < age <=100):
        print('Вітаю! Ви можете брати участь у змаганні')
    elif (sex == 'f') & (18 < age <=100):
        print('Вітаю! Ви можете брати участь у змаганні')
    else:
        print('Вибачте, на жаль, ваш не підходить для участі в змаганнях')

except SexError:
    print('Стать введено невірно.')
except ValueError:
    print('Вік введено невірно.')

```

### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import math
import statistics

#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
list_with_random_numbers
    
```

```python
def min_and_max(numbers):
    """
    Функція, яка знаходить мінімальне та максимальне значення зі списку
    """
    min_num = numbers[0]
    max_num = numbers[0]
    
    for i in numbers:
        if i < min_num:
            min_num = i
        if i > max_num:
            max_num = i 
    return min_num, max_num

def my_mean(numbers):
    """
    Функція, яка знаходить середнє значення в послідовності
    """
    mean_value = sum(numbers) / len(numbers)
    return mean_value

def my_median(numbers):
    """
    Функція, яка знаходить медіану послідовності
    """
    numbers.sort()
    if len(numbers) % 2 == 0:
        l_index = int(len(numbers) / 2) - 1
        r_index = int(len(numbers) / 2)
        median = (numbers[l_index] + numbers[r_index]) / 2
    else:
        median = len(numbers) // 2
    return median

def number_frequency(numbers):
    """
    Функція, яка знаходить частоту кожної цифри в послідовності
    """
    num_dict = {}
    
    for i in numbers:
        if i in num_dict:
            num_dict[i] += 1
        else:
            num_dict[i] = 1
            
    return num_dict
            
def mean_square_dev(numbers): 
    """
    Функція, яка знаходить середнє квадратичне відхилення в послідовності
    """
    mean = sum(numbers) / len(numbers)
    list_square_dev = list(map(lambda x: (x - mean) ** 2 , numbers))
    mean_sq_dev = math.sqrt(sum(list_square_dev) / len(list_square_dev))
    return mean_sq_dev     
```

```python
# Медіана на середнє квадратичне відхилення з бібліотекою statistics
stat_dev = statistics.stdev(list_with_random_numbers)
stat_median = statistics.median(list_with_random_numbers)
```

```python
print(f'Мінімальне значення в послідовності: {min_and_max(list_with_random_numbers)[0]}')
print(f'Mаксимальне значення в послідовності: {min_and_max(list_with_random_numbers)[1]}')
print(f'Середнє значення: {my_mean(list_with_random_numbers)}')
print(f'Медіана: {my_median(list_with_random_numbers)}')
print(f'Медіана (math): {stat_median}')      
print(f'Частота кожної цифри в послідовності: {number_frequency(list_with_random_numbers)}')
print(f'Середнє квадратичне відхилення: {mean_square_dev(list_with_random_numbers)}')
print(f'Середнє квадратичне відхилення (statistics): {stat_dev}')
```
