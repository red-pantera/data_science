У файлі "Homework-06-base.txt" мітиться інформація про товарні позиції деякого інтернет-магазину, а саме:
	Ариткул - унікальний номер товарної позиції
	Найменування - найменування товару
	Наявність - чи є товарна позиція в даний момент
	Ціна - вартість товарної позиції

#### Завдання 1
	Вам потрібно отримати таку інформацію:
		- кількість товарних позицій у файлі
		- кількість товарних позицій, що в наявності
    	- ціна та назва найдорожчої товарної позиції 

```python
DB = {}

f = open('Homework_06-base.txt', encoding='UTF-8')

for line in open('Homework_06-base.txt', encoding='UTF-8'):
    temp = line.split('|')
    part = temp[0]
    name = temp[1]
    on_stock = temp[2]
    price = int(temp[3].split()[0])
    DB[part] = [name, on_stock, price]   
f.close()
DB
```

```python
in_stock = len(list(filter(lambda key: DB[key][1] == 'В наличии', DB)))
```

```python
print(f'Кількість товарних позицій у файлі - {len(DB)}')

print(f'Кількість товарних позицій, що в наявності - {in_stock}')
```

```python
price = -1
name = None

for part in DB:
    if DB[part][2] > price:
        price = DB[part][2]
        name = DB[part][0]

        ціна та назва найдорожчої товарної позиції 
```

```python
print(f'Ціна та назва найдорожчої товарної позиції - {name}    {price}')
```

```python
import json

prod_keys = ('part', 'name', 'status', 'price')  # ключі для json
product_id = input('Введіть артикул запчстини: ')

with open('data.json', 'r') as f:
    db = json.loads(f.read())
    for i in db:
        if i == product_id:
            product = [i]
            product.extend(data[i])
            searched_product = {prod_keys[j]: product[j] for j in range(4)}
            break
print(json.dumps(searched_product, ensure_ascii = False))  
            
```
