```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
df.describe()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))

ax[0,0].plot(df['a'])
ax[0,1].plot(df['b'])
ax[1,0].plot(df['c'])

ax[1,1].plot(df.index, df.index);
```

Створити візуалізацію, аналогічно рисунку 

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(df['a']);
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
fig, ax = plt.subplots(2,1, figsize=(15,8))
ax[0].plot(df['a'])
ax[1].plot(df['b']);
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(df[['a', 'b']]);
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`
- задайте стиль 'darkgrid' за допомогою команди `sns.set_style`

```python
sns.set_style("darkgrid")

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(df[['a', 'b']]);
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(df['a'], color='r', ls='-.')
ax.plot(df['b'], color='orange', lw=10)
ax.plot(df['c'], color='yellow', lw=1, marker='o');
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
fig, ax = plt.subplots(3,1, figsize=(15,8))
ax[0].plot(df['a'], label='Line A')
ax[0].legend(loc=4)
ax[1].plot(df['b'], label='Line B')
ax[1].legend(loc=6)
ax[2].plot(df['c'], label='Line C')
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
fig, ax = plt.subplots(figsize=(15,8))
plt.bar(df.index, df['a'], color='tomato', width=0.8, alpha=0.5, label='a')
ax.legend(loc=1);
```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(df['a'], marker="^")
ax.set_title('This is the Title')
ax.set_xlabel('This is the X Axis', fontsize=20)
ax.set_ylabel('This is the Y Axis')
plt.xticks(df.index);
```
